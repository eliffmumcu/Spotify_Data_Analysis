```python
#importing the necessary libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

# Reading the Steaming_History CSV file into a pandas DataFrame
df = pd.read_json('StreamingHistory0.json') 
```

```python
# To create a new column 'Playtime'
df["PlayTime"] = pd.to_datetime(df["endTime"]) 

# The year the song was played
df['year'] = pd.DatetimeIndex(df['PlayTime']).year 

 # The month the song was played
df['month'] = pd.DatetimeIndex(df['PlayTime']).month

# The date of the day the song was played
df['day'] = pd.DatetimeIndex(df['PlayTime']).day 

 # The weekday the song was played(numerical)
df['weekday'] = pd.DatetimeIndex(df['PlayTime']).weekday

# The time the song was played
df['time'] = pd.DatetimeIndex(df['PlayTime']).time 

# The hour the song was played
df['hours'] = pd.DatetimeIndex(df['PlayTime']).hour 

# The weekday in name form the song was played
df['day-name'] = df['PlayTime'].apply(lambda x: x.day_name()) 
```

```python
# Creating a new column 'Count' to represent the number of times each song was played
df['Count'] = 1

# Grouping by song and aggregating the counts
song_counts = df.groupby(['trackName', 'artistName']).agg({'Count': 'sum'}).reset_index()

# Assuming 'song_counts' is your DataFrame with song counts
songs_played_more_than_10_times = song_counts[song_counts['Count'] > 10]

# Viewing the result
print(songs_played_more_than_10_times)
#there are 250 songs that have been streamed more than 10 times 
```

```python
# Assuming 'song_counts' is your DataFrame with song counts
top_ten_songs = song_counts.sort_values(by='Count', ascending=False).head(10)

# Viewing top ten streamed songs
print(top_ten_songs)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(song_counts['Count'], bins=60, color='skyblue', edgecolor='black')
plt.title('Histogram of Song Play Counts')
plt.xlabel('Number of Plays')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(song_counts['Count'], bins=10, range=(1, 10), color='skyblue', edgecolor='black')
plt.title('Histogram of Song Play Counts (1 to 10 times)')
plt.xlabel('Number of Plays')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
#to understand the previous data better I decieded not to look at the outlier values and compared the streaming times of 1 to 10 within themselves.
```

```python
df['day_of_week'] = pd.DatetimeIndex(df['PlayTime']).dayofweek

# Map days of the week to 'Weekday' or 'Weekend'
df['day_type'] = df['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Calculate total plays for each day type
total_plays = df['day_type'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='day_type', data=df, palette='coolwarm')
plt.title('Number of Plays on Weekdays vs. Weekends')
plt.xlabel('Day Type')
plt.ylabel('Number of Plays')

# Add percentage annotations
for i, count in enumerate(total_plays):
    percentage = (count / df.shape[0]) * 100
    plt.text(i, count + 5, f'{percentage:.2f}%', ha='center', va='bottom')

plt.show()
```

```python
# Count the number of songs played on each date
songs_by_date = df.groupby(df['PlayTime'].dt.date)['Count'].sum().reset_index()

# Plotting the bar graph
plt.figure(figsize=(200, 40))
sns.barplot(x='PlayTime', y='Count', data=songs_by_date, palette='viridis')
plt.title('Number of Songs Played by Date')
plt.xlabel('Date')
plt.ylabel('Number of Songs Played')
plt.xticks(rotation=45)
plt.show()
```

```python
df['PlayTime'] = pd.to_datetime(df['endTime'])

# Extracting the date and setting it as the index
df.set_index('PlayTime', inplace=True)

# Resampling the data to get a daily count of streams
daily_streams = df.resample('D').size()

# Plotting the user engagement over time
plt.figure(figsize=(12, 6))
daily_streams.plot(marker='o', linestyle='-', color='b')
plt.title('User Engagement Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Streams per Day')
plt.grid(True)
plt.show()
```

```python
df['month'] = pd.DatetimeIndex(df['endTime']).month


# Define a function to map months to seasons
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

# Apply the function to create a new 'Season' column
df['Season'] = df['month'].apply(get_season)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x='Season', data=df, palette='viridis')
plt.title('Number of Plays by Season')
plt.xlabel('Season')
plt.ylabel('Number of Plays')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.title('Distribution of Plays by Season')
plt.pie(df['Season'].value_counts(), labels=df['Season'].value_counts().index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.show()
#More songs were streamed in winter and fall compared to spring and summer
```

```python
df['day_of_month'] = pd.DatetimeIndex(df['endTime']).day

# Plotting a countplot for days of the month
plt.figure(figsize=(12, 6))
sns.countplot(x='day_of_month', data=df, palette='plasma')
plt.title('Number of Plays by Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Number of Plays')
plt.show()
```

```python
df['PlayTime'] = pd.to_datetime(df['endTime'])

# Define a function to map hours to 'day' or 'night'
def categorize_day_night(hour):
    if 6 <= hour < 18:
        return 'day'
    else:
        return 'night'

# Apply the function to create a new 'day-night' column
df['day-night'] = df['PlayTime'].dt.hour.apply(categorize_day_night)

# Plot the count by 'day-night'
plt.figure(figsize=(10, 6))
sns.countplot(x='day-night', data=df, palette='twilight')
plt.title('Number of Plays by Day-Night')
plt.xlabel('Day-Night')
plt.ylabel('Number of Plays')
plt.show()
```
